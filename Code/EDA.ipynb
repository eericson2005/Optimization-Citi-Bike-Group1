```python
import pandas as pd
import json

# Sample first 10,000 rows (or all rows if fewer)
n = min(10_000, len(df))
s = df.head(n).copy()

# Parse slots from the JSON string in 'extra'
def get_slots(x):
    if pd.isna(x):
        return None
    try:
        return json.loads(x).get("slots")
    except Exception:
        return None

s["slots"] = s["extra"].apply(get_slots)

# Check equality (ignore rows where slots is missing)
mask = s["slots"].notna() & s["bikes"].notna() & s["free"].notna()
s_check = s.loc[mask].copy()
s_check["bikes_plus_free"] = s_check["bikes"] + s_check["free"]
s_check["diff"] = s_check["slots"] - s_check["bikes_plus_free"]

print("rows checked:", len(s_check))
print("exact matches (diff==0):", (s_check["diff"] == 0).sum())
print("mismatch count:", (s_check["diff"] != 0).sum())
print("\ndiff value counts (top 10):")
print(s_check["diff"].value_counts().head(10))

# Show a few mismatches (if any)
mism = s_check[s_check["diff"] != 0].head(20)
if len(mism):
    print("\nexample mismatches:")
    print(mism[["timestamp", "nuid", "name", "bikes", "free", "slots", "diff", "extra"]])
```

```python
import pandas as pd

# ---- 1) Load trip history CSV and parse timestamps ----
# Change filename + column names if yours differ
trips_path = "./Data/202501-citibike-tripdata_1.csv"

trips = pd.read_csv(
    trips_path,
    parse_dates=["started_at", "ended_at"],   # timestamps in the trip file
    infer_datetime_format=True
)

# If you want everything in UTC (recommended), localize/convert as needed:
# If timestamps are already timezone-aware, this will keep them consistent.
# If they are naive, assume NYC local time and convert to UTC.
if trips["started_at"].dt.tz is None:
    trips["started_at"] = trips["started_at"].dt.tz_localize("America/New_York").dt.tz_convert("UTC")
    trips["ended_at"]   = trips["ended_at"].dt.tz_localize("America/New_York").dt.tz_convert("UTC")
else:
    trips["started_at"] = trips["started_at"].dt.tz_convert("UTC")
    trips["ended_at"]   = trips["ended_at"].dt.tz_convert("UTC")

print(trips[["started_at", "ended_at"]].head())
print(trips.dtypes[["started_at", "ended_at"]])


# ---- 2) Load CityBikes station snapshots Parquet and parse timestamps ----
snap_path = "./Data/202501-citi-bike-nyc-stats.parquet"

snaps = pd.read_parquet(snap_path)

# Common column name is 'timestamp'. If yours differs, change it here.
snaps["timestamp"] = pd.to_datetime(snaps["timestamp"], utc=True)

print(snaps[["timestamp", "nuid", "bikes", "free"]].head())
print(snaps.dtypes[["timestamp"]])
```
